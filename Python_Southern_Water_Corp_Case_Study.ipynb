### Step 6: <span style="color:green">Create Quartiles</span> 

i) Create two new variables called Q1 and Q3. q1 should contain the 25th percentile for all columns in the dataframe while Q3 should contain the 75th percentile for all the columns in the dataframe.

ii) Calculate the interquartile range **(IQR = Q3 - Q1)** for all columns in the dataframe and print it to the screen.

## Please put your code here

```python
#(i) Create two new variables called Q1 and Q3
Q1 = Merged_df.quantile(0.25)
print(Q1)

Q3 = Merged_df.quantile(0.75)
print(Q3)

#(ii) Calculate the interquartile range (IQR = Q3 - Q1) for all columns in the dataframe and print it to the screen.
IQR = Q3-Q1
print(IQR)
```

### Step 7: <span style="color:green">Identify Outliers</span> 

How many outliers do you have? What will happen to your dataset if you remove them all? Let's find out!

i) Calculate how many entries you currently have in the original dataframe.

ii) Using the quartiles and IQR previously calculated, identify the number of entries you'd have if you were to remove the outliers.

ii) Find the proportion of outliers that exist in the dataset.

Ensure your dataframe doesn't include the attribute TIMEFRAME - if it does, please drop this attribute for now.


## Please put your code here

```python
#i. The number of entries in the original dataframe.
Merged_df = pd.concat(frame)
df = Merged_df.drop('TIMEFRAME', axis=1)
print ("i. When we have NOT removed any outliers from the dataset, we have " + str(len(df.index))  + " entries/rows.") 

#ii. The number of entries if you were to remove the outliers.
removed_outliers = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)] #Bitwise operators!!
print ("ii. When we removed outliers from the dataset, we have " + str(len(removed_outliers.index))  + " entries/rows.")

#iii. The proportion of outliers that exist in the dataset.
outliers = ((len(df.index) - len(removed_outliers.index))/len(df.index))*100
print ("iii. The proportion of outliers which exist when compared to the original dataframe is " + str(round(outliers,2)) + '%.')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#Answer for step i): 
#Below is the first part of the code
Merged_df2 = pd.concat(frame)
Merged_df2 = Merged_df2.dropna()
dependentVar = Merged_df2['PUMP FAILURE (1 or 0)']
#Merged_df2['TIMEFRAME'] = pd.to_datetime(Merged_df2['TIMEFRAME'], format="%d/%m/%Y %H:%M:%S", infer_datetime_format=True )
#Merged_df2.set_index('TIMEFRAME', inplace=True)
filt = ['SURJEK_FLOW_METER_1', 'SURJEK_FLOW_METER_2', 'ROTATIONAL_PUMP_RPM',
       'SURJEK_PUMP_TORQUE', 'MAXIMUM_DAILY_PUMP_TORQUE',
       'SURJEK_AMMONIA_FLOW_RATE', 'SURJEK_TUBE_PRESSURE',
       'SURJEK_ESTIMATED_EFFICIENCY']

#----write your code below------
X = Merged_df2[filt] #independent factors
Y = Merged_df2['PUMP FAILURE (1 or 0)'] #dependent factor
X = sm.add_constant(X) # adding a constant term to the predictor for y-int

model = sm.OLS(Y,X)
results = model.fit() 
results.summary()
```

RESULTS:
What is the R Squared for the model and what does this signify?
'R-squared' reflects the fit of the model with the higher value within the range of 0-1 indicating a better fit. This OLS regression result shows R-squared of 26% suggesting a low fit.

-The 'const' coef(ficient) is the y-intercept and it is 0.1293.

-The coef(ficient) for each of the factors (i.e. Surjek Flow Meter 1) reflects the change in the output of Y given one unit change in the factor with everything else held constant.


```python
#answer for step ii): 
#Below is the first part of the code
Merged_df2 = pd.concat(frame)
Merged_df2 = Merged_df2.dropna()
dependentVar = Merged_df2['PUMP FAILURE (1 or 0)']
filt = ['SURJEK_FLOW_METER_1', 'SURJEK_FLOW_METER_2', 'ROTATIONAL_PUMP_RPM',
       'SURJEK_PUMP_TORQUE', 'MAXIMUM_DAILY_PUMP_TORQUE',
       'SURJEK_AMMONIA_FLOW_RATE', 'SURJEK_TUBE_PRESSURE',
       'SURJEK_ESTIMATED_EFFICIENCY']

Rolling_df2 = Merged_df2[filt].rolling(30).std() 

#----write your code below------
#ii) Repeat i) but this time use the rolling standard deviation variables you created previously. What is the R Squared for the model and what does this signify?
X2 = Rolling_df2 
Y2 = Merged_df2['PUMP FAILURE (1 or 0)'] 
X2 = X2.fillna(0)
X2 = sm.add_constant(X2) 
print(X2.head())

model2 = sm.OLS(Y2,X2) 
results2 = model2.fit()
results2.summary()

#RESULT: The rolling 30-point standard deviation seems to be a better fit given R-squared is 63% vs 26% previously.
```
